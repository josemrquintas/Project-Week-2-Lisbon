```python
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
```

```python
df=pd.read_csv('query.csv')
```

```python
def get_freguesia(x):
    new_list=x.split(', ')
    freguesia=new_list[new_list.index('Lisboa')-1]
    return freguesia
```

```python
geolocator = Nominatim(user_agent="Tiago",timeout=None)

df['X'] = df['X'].astype(str)
df['Y']= df['Y'].astype(str)
df['Coordinates'] = df[['Y', 'X']].apply(lambda x: ','.join(x), axis=1)
df['Morada']=[str(geolocator.reverse(df['Coordinates'][row])) for row in range(len(df['X']))]
df['Freguesia']=df['Morada'].apply(get_freguesia)

df.head(30)
    
```

```python
df2=df[['NOME','Freguesia']]

#final_result=(df2.groupby(['Freguesia']).count()).rename(columns={"NOME": "police_station_freq"})
final_result=df2.groupby(['Freguesia']).agg(['count'])
final_result
```
