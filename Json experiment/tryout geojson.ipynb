```python
import pandas as pd 
import numpy as np
import json
```

```python
data = pd.read_json('Ciclovias.geojson')
data1 = data['features']
data1
```

```python
data2 = dict(data1)

data3 = dict(data2)

print(data2.keys()) #- index
print(data2[0]['properties'].keys()) #- columns
```

```python
data3
```

```python
from pandas.io.json import json_normalize

df = json_normalize(data2["features"])

coords = 'properties.geometry.coordinates'

df2 = (df[coords].apply(lambda r: [(i[0],i[1]) for i in r[0]])
           .apply(pd.Series).stack()
           .reset_index(level=1).rename(columns={0:coords,"level_1":"point"})
           .join(df.drop(coords,1), how='left')).reset_index(level=0)

df2[['lat','long']] = df2[coords].apply(pd.Series)

df2
```
