```python
import numpy as np
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
```

```python
secundarios = pd.read_csv('secundarios.csv')
secundarios.head()
```

```python
geolocator = Nominatim(user_agent="josemrquintas", timeout = None)


secundarios['MORADA'] = [(geolocator.reverse(str(secundarios['Y'][i]) + ', ' + str(secundarios['X'][i]))) for i in range(len(secundarios['X']))]

secundarios.head()
```

```python
secundarios['MORADA1'] = [str(secundarios['MORADA'][x]).split(',') for x in range(len(secundarios['MORADA']))]

secundarios.tail(20)
```

```python
list_sec_neighbourhoods = []

for p in range(len(secundarios['MORADA1'])): 
    for x in range( len( secundarios['MORADA1'][p] ) -1 ):
        if secundarios['MORADA1'][p][x+1] == ' Lisboa': 
            list_sec_neighbourhoods.append((str((secundarios['MORADA1'][p][x]))).lstrip())


secundarios['FREGUESIA'] = list_sec_neighbourhoods


secundarios.head()
```

```python
secundarios_filtered = secundarios[['X','Y','INF_NOME', 'INF_DESCRICAO', 'MORADA1','FREGUESIA']]

secundarios_filtered.head()
```

```python
secundarios_filtered.to_csv('secundarios_filtered.csv',index = False)
```
