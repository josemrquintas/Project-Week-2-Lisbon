```python
import numpy as np
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
import time
```

```python
universities = pd.read_csv('superior.csv')
universities
```

```python
geolocator = Nominatim(user_agent="josemrquintas")

uni_adresses =[]

for i in range(len(universities['X'])) : 
    time.sleep(0.5)
    uni_adresses.append(geolocator.reverse(str(universities['Y'][i]) + ', ' + str(universities['X'][i])))

universities['MORADA'] = uni_adresses

universities.head()
```

```python
universities['MORADA1'] = [str(universities['MORADA'][x]).split(',') for x in range(len(universities['MORADA']))]

universities.tail(20)
```

```python
list_uni_neighbourhoods = []

for p in range(len(universities['MORADA1'])): 
    for x in range( len( universities['MORADA1'][p] ) -1 ):
        if universities['MORADA1'][p][x+1] == ' Lisboa': 
            list_uni_neighbourhoods.append((str((universities['MORADA1'][p][x]))).lstrip())


universities['FREGUESIA'] = list_uni_neighbourhoods


universities.head()
```

```python
universities_filtered = universities[['X','Y','INF_NOME', 'INF_DESCRICAO', 'MORADA1','FREGUESIA']]

universities_filtered.head()
```

```python
universities_filtered.to_csv('universities_filtered.csv',index = False)
```
