```python
parque_filt = parque[['X', 'Y', 'MORADA', 'DESIGNACAO','TIPOLOGIA','Shape__Area','Shape__Length']]
```

```python
parque_filt
```

```python
import geopy
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="abcdded", timeout=None)
#location = geolocator.reverse("38.716381, -9.131562")
```

```python
###MVP####


list1 = [(geolocator.reverse(str(parque_filt['Y'][i]) + ', ' + str(parque_filt['X'][i]))) for i in range(len(parque_filt['X']))]
```

```python
parque_filt['FREGUESIA'] = list1
```

```python
parque_filt
```

```python
###MVP####

parque_filt['FREGUESIA'] = [(str(parque_filt['FREGUESIA'][p]).split(','))[z].lstrip() for p in range(len(parque_filt['FREGUESIA'])) for z in range( len( str(parque_filt['FREGUESIA'][p]).split(',')) -1 ) if (str(parque_filt['FREGUESIA'][p]).split(','))[z+1] == ' Lisboa']
```

```python
parque_filt
```

```python
# drop null

null_cols = parque_filt.isnull().sum()
null_cols

```

```python
#Drop X and Y and TIPOLOGIA

parque_filt_drop = parque_filt.drop(['X','Y','TIPOLOGIA'], axis =1)

parque_filt_drop
```

```python

column_order = ['FREGUESIA', 'Shape__Area', 'Shape__Length','DESIGNACAO'] #IT REORDER AND DELETE THE ONES WE DONT ADD IN THE LIST
parque_filt_drop = parque_filt_drop[column_order]
parque_filt_drop
```
